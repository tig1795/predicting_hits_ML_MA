```python
result = df.sentiment.str.split(',', expand=True)
```

```python
result
```

```python
result = result.rename({0: 'negative', 1: 'neutral', 2: 'positive', 3: 'compound'}, axis=1)
```

```python
result
```

```python
for i, row in result.iterrows():
    result.at[i, 'negative'] = row.negative.replace('{',' ').replace(' '' ',' ').replace(':',' ').lower()
result
```

```python
result['neutral'] = result['neutral'].str.replace('neu',' ')
```

```python
result['neutral'] = result['neutral'].str.replace(':',' ')
```

```python
result['neutral'] = result['neutral'].str.replace(" ' ", " ")
```

```python
result['neutral'] = result['neutral'].str.replace(" ' ", " ")
```

```python
result.head(5)
```

```python
result = result.drop(["compound"], axis=1)
```

```python
result
```

```python
result['negative'] = result['negative'].astype('float')
```

```python
result['neutral'] = result['neutral'].astype('float')
```

```python
result['positive'] = result['positive'].astype('float')
```

```python
result
```

```python
step = df.join(result)
```

```python
step.head(5)
```

```python
step.shape
```

```python
import pandas as pd

# create a new column to hold the name of the column with the maximum value
step['max_sentiment'] = step[['negative', 'neutral', 'positive']].idxmax(axis=1)

step.head(5)
```

```python
step.shape
```

```python
step_hit = step[step['Hit'] == 'Yes']
```

```python
step_no_hit = step[step['Hit'] == 'No']
```

```python
s1 = step_hit['max_sentiment'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(15,8))

ax=s1.plot.bar(width=.9, color=('blue', 'orange', 'green', 'red', 'purple', 'brown', 'grey')) 
for i, v in s1.reset_index().iterrows():
    ax.text(i, v.max_sentiment + 0.2 , v.max_sentiment, color='black')

ax.set_xlabel(len(step_hit.max_sentiment))
plt.title("Ergebnis Sentimental Analysis für die Hit Songs")
#plt.xlabel(len())
plt.ylabel('Häufigkeit')
#pltcountplotfig("")
#plt.savefig('')
```

```python
s = step_no_hit['max_sentiment'].value_counts().head(10)
```
