## Wordcloud for genre tag of hit songs

```python
# Select first the hit genres

hitGenres = list(let[let["category"] == "Hit"]["Genres"])
```

```python
len(hitGenres)
```

```python
# strip and replace bits in the Genre strings
hitGenres_long = []
for thisTrack in hitGenres:
    thisTrackGenres = thisTrack.replace('\'', '').replace('[', '').replace(']', '').replace(', ', ',').split(',')
    hitGenres_long.extend(thisTrackGenres)
```

```python
from collections import Counter

hitGenres_count = Counter(hitGenres_long)
```

```python
from wordcloud import WordCloud
import matplotlib
import matplotlib.pyplot as plt

hitWordcloud = WordCloud(stopwords='unknown', 
        height=700,
        width=1400,
        min_font_size=4,
        colormap=matplotlib.cm.inferno,
        background_color='white').generate_from_frequencies(hitGenres_count)
```

```python
hitWordcloud.to_file('Plots/wordcloud_hit_genres_new.png')
```

# Audio features

## Statistics

```python
audiofeatures = let[['category', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
 'instrumentalness', 'liveness', 'valence', 'tempo', 'key', 'mode']]
```

```python
audiofeatures.head(5)
```

```python
from collections import Counter

Counter(audiofeatures.category)
```

```python
hitValues = audiofeatures[audiofeatures['category']=='Hit']
nohitValues = audiofeatures[audiofeatures['category']=='no Hit']
```

```python
hitValues.shape
```

```python
nohitValues.shape
```

```python
features =['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
'valence', 'tempo']
```

```python
import pingouin as pg

statHolder = pd.DataFrame()

for thisFeature in features:
    output = pg.ttest(hitValues[thisFeature],
                     nohitValues[thisFeature],
                     paired=False,
                     alternative='two-sided',
                     correction='auto')
    output['feature'] = thisFeature
    statHolder = statHolder.append(output, ignore_index=True)
```

```python
statHolder.head(5)
```

```python
bonferroniCorrected = pg.multicomp(list(statHolder['p-val']), alpha=.01, method='bonf')
```

```python
# extract the corrected p-values
statHolder['p-corr'] = bonferroniCorrected[1]

# extract a logical array if test is significant
statHolder['significant'] = bonferroniCorrected[0]

# write the statistics to file
statHolder.to_csv('Data/hit_nonhit_Comparison_new.csv', encoding='UTF-8')
```

```python
statHolder.shape
```

## Raincloud plots

```python
features =[
'danceability',
'energy',
'loudness',
'speechiness',
'acousticness',
'instrumentalness',
'liveness',
'valence',
'tempo'
]
```

```python
import ptitprince as pt

# make the plots in a loop
for thisFeature in features:
    # Make a name for the figure
    saveName = 'Plots/hits_non_hits_new-' + thisFeature + '.png'
    # draw the plot
    f, ax = plt.subplots(figsize=(7,5), dpi=300)
    pt.RainCloud(x='category', y=thisFeature, data=audiofeatures, palette='Set2', bw=.2, 
 width_viol=.6, move=.2, ax=ax, orient='h', point_size=.35,
 box_showfliers = False)
    # Letâ€™s try to draw statistics in the plot as well
    # get stats for this feature
    thisStat =       statHolder[statHolder['feature']==thisFeature].reset_index()
    # check if significant
    isSignificant = thisStat['significant']
    signifStar = ''
    if isSignificant[0]:
        # you can use this to add a significance star somewhere
        signifStar = '*' 
    # get p-value
    pVal = thisStat['p-corr']
    # if pVal is below 0.0001, then print as p < 0.001
    if float(pVal.round(3))<=0.0:
        pVal = '< 0.001'
    else:
        pVal = '= ' + str(float(pVal.round(3)))
    # get Cohenâ€™s d-value
    dVal = thisStat['cohen-d']
    # make a string with the statistics
    statText = signifStar + 'P ' + pVal + ', Cohen\â€™s D = ' +    str(float(dVal.round(3)))
    # add title
    plt.suptitle(thisFeature.capitalize())
    # add statistics
    plt.title(statText)
    plt.xlabel('Value')
    plt.ylabel('')
    # save the figure
    plt.savefig(saveName, bbox_inches='tight')
```
