# Anova Test

```python
import pandas as pd
from scipy.stats import f_oneway

# Load the data into a pandas DataFrame
df = pd.read_csv("../../data/audiofeatures/hits_no_hits_new.csv", index_col=0)

df.head(5)
```

```python
# Replace "Hit" with 1, "no Hit" with 0, and "None" with NaN
df["category"] = df["category"].replace({"Hit": 2, "no Hit": 1, "None": 0})
```

```python
df
```

```python
import collections

c = collections.Counter(df.category)

c
```

```python
# Split the data into hit songs and no hit songs
hits = df[df["category"] == 'Hit']["danceability"]
no_hits = df[df["category"] == 'no Hit']["danceability"]

# Perform the ANOVA test
f_value, p_value = f_oneway(hits, no_hits)

# Print the results
print("F-value:", f_value)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference between the audio features of hit songs and no hit songs.")
else:
    print("There is not a significant difference between the audio features of hit songs and no hit songs.")
```

```python
import numpy as np
from scipy.stats import f_oneway
import matplotlib.pyplot as plt

# Split the data into hit songs and no hit songs
hits = df[df["category"] == 'Hit']["danceability"]
no_hits = df[df["category"] == 'no Hit']["danceability"]

# Perform the ANOVA test
f_value, p_value = f_oneway(hits, no_hits)

# Calculate the effect size (eta-squared)
sst = np.sum((df["danceability"] - np.mean(df["danceability"])) ** 2)
ssm = (np.sum((hits - np.mean(df["danceability"])) ** 2) + np.sum((no_hits - np.mean(df["danceability"])) ** 2))
n_groups = len(df["category"].unique())
n_total = len(df["category"])
eta_sq = ssm / sst / (n_groups - 1) * (n_total - n_groups)

# Visualize the results with a box plot
plt.boxplot([hits, no_hits], labels=["Hit", "no Hit"])
plt.title("Distribution of audio features for hit songs and no hit songs")
plt.xlabel("Song category")
plt.ylabel("Audio feature value")
plt.show()

# Print the results
print("F-value:", f_value)
print("P-value:", p_value)
print("Eta-squared:", eta_sq)

if p_value < 0.05:
    print("There is a significant difference between the audio features of hit songs and no hit songs.")
else:
    print("There is not a significant difference between the audio features of hit songs and no hit songs.")
```

# Correlation analysis

```python
import pandas as pd
from scipy.stats import pearsonr

# Filter the DataFrame to include only hit and no hit songs
df_hits = df[df["category"] == 2]
df_no_hits = df[df["category"] == 1]

# Calculate the correlation coefficient and p-value between danceability and hit songs
corr_coef_hits, p_value_hits = pearsonr(df_hits["danceability"], df_hits["category"])

# Calculate the correlation coefficient and p-value between danceability and no hit songs
corr_coef_no_hits, p_value_no_hits = pearsonr(df_no_hits["danceability"], df_no_hits["category"])

# Print the results
print(f"Correlation coefficient between danceability and hit songs: {corr_coef_hits:.3f}")
print(f"P-value: {p_value_hits:.3f}")

print(f"Correlation coefficient between danceability and no hit songs: {corr_coef_no_hits:.3f}")
print(f"P-value: {p_value_no_hits:.3f}")
```
