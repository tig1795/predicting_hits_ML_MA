## Statistics

Es gibt viele Mglichkeiten, einen statistischen Vergleich zwischen zwei Gruppen durchzufhren. Wir werden hier nicht im Detail darauf eingehen, sondern stattdessen zeigen, wie man einen einfachen T-Test fr einige der Klangmerkmale durchfhrt. Dies gibt uns einen Hinweis darauf, ob es einen statistisch signifikanten Unterschied zwischen den Klangmerkmalen von hits und no hits gibt.

Zunchst werden wir unseren Datensatz so unterteilen, dass er nur die Kategorie (Hit oder no Hit) und die Audiomerkmale enthlt:

```python
audiofeatures = daf[['category', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
 'instrumentalness', 'liveness', 'valence', 'tempo', 'key', 'mode']]
```

```python
audiofeatures.head(5)
```

```python
from collections import Counter

Counter(audiofeatures.category)
```

Um unsere Berechnungen etwas zu vereinfachen, werden wir dies in zwei Dataframes aufteilen, einen fr Hits und einen fr no Hits:

```python
hitValues = audiofeatures[audiofeatures['category']=='Hit']
nohitValues = audiofeatures[audiofeatures['category']=='no Hit']
```

```python
hitValues.shape
```

```python
nohitValues.shape
```

Anschlieend erstellen wir eine Liste mit den Audiofunktionen, die wir vergleichen mchten:

```python
features =['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
'valence', 'tempo']
```

Der T-Test wird mit der ttest-Funktion von Pingouin durchgefhrt. Diese Funktion gibt einen Pandas-Datenrahmen mit allen relevanten Informationen aus. Wir fhren unsere eigentliche Berechnung in einer Schleife durch, um alle Merkmale zu durchlaufen.

```python
import pingouin as pg

statHolder = pd.DataFrame()

for thisFeature in features:
    output = pg.ttest(hitValues[thisFeature],
                     nohitValues[thisFeature],
                     paired=False,
                     alternative='two-sided',
                     correction='auto')
    output['feature'] = thisFeature
    statHolder = statHolder.append(output, ignore_index=True)
```

```python
statHolder.head(5)
```

Ich werde hier nicht auf die eigentlichen Statistiken eingehen, aber die Kurzfassung ist, dass dies zwei Variablen ausgibt, die uns interessieren. statHolder hat nun eine Spalte namens "p-val", die die p-Werte enthlt, und eine Spalte namens "cohen-d", die Cohen's D (ein Ma fr die Effektgre) enthlt.

Im Allgemeinen werden bei einem so groen Datensatz selbst kleine Unterschiede zwischen den Verteilungen als statistisch signifikant angesehen. Dies bedeutet jedoch nicht immer, dass ein Unterschied signifikant ist. Die Effektgre, in diesem Fall Cohen's D, gibt einen Hinweis auf das Ausma des Unterschieds.

Wir haben gerade eine Reihe von T-Tests durchgefhrt, und das bedeutet normalerweise, dass wir fr Mehrfachvergleiche korrigieren sollten. Hier werden wir die Bonferroni-Korrektur verwenden.

Pingouin verfgt ber eine Funktion, die Korrekturen vornimmt:

```python
bonferroniCorrected = pg.multicomp(list(statHolder['p-val']), alpha=.01, method='bonf')
```

Dadurch erhalten wir ein Array, das ein boolesches Wahr/Falsch enthlt, das angibt, ob ein Vergleich statistisch signifikant war (beachten Sie, dass wir Alpha, die Signifikanzschwelle, auf 0,01 gesetzt haben), sowie die korrigierten P-Werte.

Fgen wir diese in den statHolder-Datenrahmen ein und schreiben sie auf die Festplatte:

```python
# extract the corrected p-values
statHolder['p-corr'] = bonferroniCorrected[1]

# extract a logical array if test is significant
statHolder['significant'] = bonferroniCorrected[0]

# write the statistics to file
statHolder.to_csv('Data/hit_nonhit_Comparison_ohne_duplikate.csv', encoding='UTF-8')
```
