# Audio features

## Statistics

Es gibt viele Möglichkeiten, einen statistischen Vergleich zwischen zwei Gruppen durchzuführen. Wir werden hier nicht im Detail darauf eingehen, sondern stattdessen zeigen, wie man einen einfachen T-Test für einige der Klangmerkmale durchführt. Dies gibt uns einen Hinweis darauf, ob es einen statistisch signifikanten Unterschied zwischen den Klangmerkmalen von hits und no hits gibt.

Zunächst werden wir unseren Datensatz so unterteilen, dass er nur die Kategorie (Hit oder no Hit) und die Audiomerkmale enthält:

```python
audiofeatures = daf[['category', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
 'instrumentalness', 'liveness', 'valence', 'tempo', 'key', 'mode']]
```

```python
audiofeatures.head(5)
```

```python
from collections import Counter

Counter(audiofeatures.category)
```

Um unsere Berechnungen etwas zu vereinfachen, werden wir dies in zwei Dataframes aufteilen, einen für Hits und einen für no Hits:

```python
hitValues = audiofeatures[audiofeatures['category']=='Hit']
nohitValues = audiofeatures[audiofeatures['category']=='no Hit']
```

```python
hitValues.shape
```

```python
nohitValues.shape
```

Anschließend erstellen wir eine Liste mit den Audiofunktionen, die wir vergleichen möchten:

```python
features =['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
'valence', 'tempo']
```

Der T-Test wird mit der ttest-Funktion von Pingouin durchgeführt. Diese Funktion gibt einen Pandas-Datenrahmen mit allen relevanten Informationen aus. Wir führen unsere eigentliche Berechnung in einer Schleife durch, um alle Merkmale zu durchlaufen.

```python
import pingouin as pg

statHolder = pd.DataFrame()

for thisFeature in features:
    output = pg.ttest(hitValues[thisFeature],
                     nohitValues[thisFeature],
                     paired=False,
                     alternative='two-sided',
                     correction='auto')
    output['feature'] = thisFeature
    statHolder = statHolder.append(output, ignore_index=True)
```

```python
statHolder.head(5)
```

Ich werde hier nicht auf die eigentlichen Statistiken eingehen, aber die Kurzfassung ist, dass dies zwei Variablen ausgibt, die uns interessieren. statHolder hat nun eine Spalte namens "p-val", die die p-Werte enthält, und eine Spalte namens "cohen-d", die Cohen's D (ein Maß für die Effektgröße) enthält.

Im Allgemeinen werden bei einem so großen Datensatz selbst kleine Unterschiede zwischen den Verteilungen als statistisch signifikant angesehen. Dies bedeutet jedoch nicht immer, dass ein Unterschied signifikant ist. Die Effektgröße, in diesem Fall Cohen's D, gibt einen Hinweis auf das Ausmaß des Unterschieds.

Wir haben gerade eine Reihe von T-Tests durchgeführt, und das bedeutet normalerweise, dass wir für Mehrfachvergleiche korrigieren sollten. Hier werden wir die Bonferroni-Korrektur verwenden.

Pingouin verfügt über eine Funktion, die Korrekturen vornimmt:

```python
bonferroniCorrected = pg.multicomp(list(statHolder['p-val']), alpha=.01, method='bonf')
```

Dadurch erhalten wir ein Array, das ein boolesches Wahr/Falsch enthält, das angibt, ob ein Vergleich statistisch signifikant war (beachten Sie, dass wir Alpha, die Signifikanzschwelle, auf 0,01 gesetzt haben), sowie die korrigierten P-Werte.

Fügen wir diese in den statHolder-Datenrahmen ein und schreiben sie auf die Festplatte:

```python
# extract the corrected p-values
statHolder['p-corr'] = bonferroniCorrected[1]

# extract a logical array if test is significant
statHolder['significant'] = bonferroniCorrected[0]

# write the statistics to file
statHolder.to_csv('Data/hit_nonhit_Comparison_ohne_duplikate.csv', encoding='UTF-8')
```

```python
statHolder.shape
```

## Raincloud plots

Wir werden jedes der Audio-Features, die wir gerade verglichen haben, durchgehen, ein Diagramm erstellen und einige Statistiken zu dem Diagramm hinzufügen. Zunächst erstellen wir eine Liste der Audio-Features, die wir darstellen wollen:

```python
features =[
'danceability',
'energy',
'loudness',
'speechiness',
'acousticness',
'instrumentalness',
'liveness',
'valence',
'tempo'
]
```

Jetzt werden wir die Handlung in einer etwas hässlichen Schleife durchführen:

```python
import ptitprince as pt

# make the plots in a loop
for thisFeature in features:
    # Make a name for the figure
    saveName = 'Plots/hits_non_hits_ohne_duplikate-' + thisFeature + '.png'
    # draw the plot
    f, ax = plt.subplots(figsize=(7,5), dpi=300)
    pt.RainCloud(x='category', y=thisFeature, data=audiofeatures, palette='Set2', bw=.2, 
 width_viol=.6, move=.2, ax=ax, orient='h', point_size=.35,
 box_showfliers = False)
    # Let’s try to draw statistics in the plot as well
    # get stats for this feature
    thisStat =       statHolder[statHolder['feature']==thisFeature].reset_index()
    # check if significant
    isSignificant = thisStat['significant']
    signifStar = ''
    if isSignificant[0]:
        # you can use this to add a significance star somewhere
        signifStar = '*' 
    # get p-value
    pVal = thisStat['p-corr']
    # if pVal is below 0.0001, then print as p < 0.001
    if float(pVal.round(3))<=0.0:
        pVal = '< 0.001'
    else:
        pVal = '= ' + str(float(pVal.round(3)))
    # get Cohen’s d-value
    dVal = thisStat['cohen-d']
    # make a string with the statistics
    statText = signifStar + 'P ' + pVal + ', Cohen\’s D = ' +    str(float(dVal.round(3)))
    # add title
    plt.suptitle(thisFeature.capitalize())
    # add statistics
    plt.title(statText)
    plt.xlabel('Value')
    plt.ylabel('')
    # save the figure
    plt.savefig(saveName, bbox_inches='tight')
```

## Erklärung Code Raincloud plots

```python
for thisFeature in features:
```

Diese Schleife durchläuft jedes Element in features, und wir können in der Variable thisFeature darauf zugreifen. In der ersten Runde der for-Schleife ist thisFeature "Tanzbarkeit".

```python
saveName = ‘Plots/MetalComparison-’ + thisFeature + ‘.png’
```

Dadurch wird eine Zeichenkette erstellt, die wir zum Speichern des Plots verwenden werden. Da thisFeature "danceability" ist, wird dies "Plots/MetalComparison-danceability.png" sein.

```python
f, ax = plt.subplots(figsize=(7,5), dpi=300)
pt.RainCloud(x=’category’, y=thisFeature, 
data=audiofeatures, palette=’Set2', bw=.2, 
width_viol=.6, move=.2, ax=ax, 
orient=’h’, point_size=.35,
box_showfliers = False)
```

So entsteht das eigentliche Raincloud-Diagramm unter Verwendung der RainCloud-Funktion aus dem Paket ptitprince. Auf der x-Achse wählen wir die Spalte "Kategorie", die entweder Hit oder no Hit ist. Auf der y-Achse wählen wir "thisFeature", das sich während der Schleife ändert. Die Daten stammen aus unserem Audiofeatures-Datenframe. Die übrigen Parameter können im RainCloud-Tutorial nachgelesen werden. Es gibt jedoch einen Trick, nämlich box_showfliers = False. Damit werden die Ausreißer im Boxplot deaktiviert.

Jetzt werden wir versuchen, einige Statistiken hinzuzufügen. Wir wollen den P-Wert und das Cohen's D zu unseren Diagrammen hinzufügen. Außerdem machen wir ein kleines Sternchen, *, um anzuzeigen, ob ein Vergleich signifikant war.

Zuerst erhalten wir die Statistiken für dieses Audio-Feature:

```python
thisStat = statHolder[statHolder[‘feature’]==thisFeature].reset_index()
```

Anschließend wird geprüft, ob sie nach der Bonferroni-Korrektur signifikant bleibt:

```python
isSignificant = thisStat[‘significant’]
signifStar = ‘’
if isSignificant[0]:
    signifStar = ‘* ‘
```

Wir haben dort eine Zeichenkette, signifStar, erstellt, die standardmäßig leer ist, aber ein * und ein Leerzeichen enthält, wenn der Vergleich signifikant war.
Wir erhalten dann den P-Wert:

```python
pVal = thisStat[‘p-corr’]
# if pVal is below 0.001, then print as p < 0.001
if float(pVal.round(3))<0.001:
    pVal = ‘< 0.001’
else:
    pVal = ‘= ‘ + str(float(pVal.round(3)))
```

Wenn der auf 3 Dezimalstellen gerundete P-Wert kleiner als 0,001 ist, wird er als "< 0,001" angezeigt. Andernfalls wollen wir ihn als "= tatsächlicher Wert" anzeigen.
Anschließend extrahieren wir den Cohen's D-Wert:

```python
dVal = thisStat[‘cohen-d’]
```

Nun formatieren wir dies in eine schöne Zeichenkette:

```python
statText = signifStar + ‘P ‘ + pVal + ‘, Cohen\’s D = ‘ + str(float(dVal.round(3)))
```

Jetzt müssen Sie nur noch Beschriftungen und Titel hinzufügen und die Abbildung speichern. Ein kleiner Trick, um einen Haupttitel und einen Untertitel zu erhalten, besteht darin, plt.suptitle für den Haupttitel und plt.title für den Untertitel zu verwenden.

```python
# add title
plt.suptitle(thisFeature.capitalize())
# add statistics
plt.title(statText)
plt.xlabel(‘Value’)
plt.ylabel(‘’)
# save the figure
plt.savefig(saveName, bbox_inches=’tight’)
```
