# Exploratory Data Analysis

target: Die Zielvariable für die Lieder. Sie kann entweder "0" oder "1" sein. 1" bedeutet, dass der Titel mindestens einmal in der wöchentlichen Liste (von Billboards) der Hot-100-Titel in diesem Jahrzehnt aufgeführt wurde und somit ein "Hit" ist. 0" bedeutet, dass der Titel ein "Flop" ist.

Die Bedingungen dafür, dass ein Titel ein "Flop" ist, lauten wie folgt:

			- Der Titel darf nicht in der Hitliste des betreffenden Jahrzehnts auftauchen.
			- Der Künstler des Titels darf nicht in der Hitliste des betreffenden Jahrzehnts auftauchen.
			- Der Titel muss einem Genre angehören, das nicht als Mainstream und/oder Avantgarde bezeichnet werden kann.
			- Das Genre des Titels darf keinen Song in der "Hit"-Liste haben.
			- Der Titel muss "US" als einen seiner Märkte haben.

# Feature Engineering

## Splitting the data into dependent and independent features

```python
X = df.drop('target',axis=1)
```

```python
X.columns
```

```python
y = df["target"]
```

## Now lets split the data in train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

Now we have split the data to train and test data

# Model Selection

```python
from xgboost import XGBClassifier
```

```python
classifier = XGBClassifier()
```

## Lets import randomized Search Cross Validation too so we can perform hyperparameter tuning on XGBoost

```python
from sklearn.model_selection import RandomizedSearchCV
```

## Lets pass a range of parameters for the Randomized Search CV to take and select the best one for us

```python
params = {
    'learning_rate':[0.05,0.10,0.15,0.20,0.25,0.30],
    'max_depth': [3,4,5,6,8,10,12,15],
    'min_child_weight':[1,3,5,7],
    'gamma':[0.0, 0.1, 0.2, 0.3, 0.4],
    'colsample_bytree': [0.3, 0.4, 0.5, 0.7]
}
```

## Passing these parameters to RandomizedSearchCV

```python
random_search = RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring="roc_auc",n_jobs=-1,cv=5,verbose=3)
```

```python
random_search.fit(x_train,y_train)
```

## Lets see what parameters were selected by the Cross Validation Process

```python
random_search.best_estimator_
```

```python
classifier= XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.7,
              enable_categorical=False, gamma=0.2, gpu_id=-1,
              importance_type=None, interaction_constraints='',
              learning_rate=0.1, max_delta_step=0, max_depth=5,
              min_child_weight=1, monotone_constraints='()',
              n_estimators=100, n_jobs=8, num_parallel_tree=1, predictor='auto',
              random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
              subsample=1, tree_method='exact', validate_parameters=1,
              verbosity=None)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report
```

```python
cross = cross_val_score(classifier,x_train,y_train,cv=10,verbose=0)
```

```python
print("Accuracy after cross validation is: {}".format(cross.mean()*100))
```

# Lets Predict the values

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt="g")
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

# KNN(K-Nearest Neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
```

## Finding the optimal K value for the KNN Classifier

```python
error_rate=[]
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

## Plotting the error rates for better visualization

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color="blue",linestyle="dashed",marker='o')
plt.title("K value vs Error Rate")
plt.xlabel("K values")
plt.ylabel("Error rate")
```

## Initializing the KNN model

```python
knn = KNeighborsClassifier(n_neighbors=37)
```

## Fitting the KNN Model

```python
knn.fit(x_train,y_train)
```

## Predicting the value for test data

```python
y_predicted = knn.predict(x_test)
```

## Generating a Classification Report

```python
print(classification_report(y_test,y_predicted))
```

```python
conf = confusion_matrix(y_test,y_predicted)
sns.heatmap(conf,annot=True,fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
