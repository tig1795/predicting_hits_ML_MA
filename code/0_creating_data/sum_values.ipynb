# Values addieren & doppelte Einträge entfernen

```python
# Group by artist and title columns, and sum the values column
grouped_df = df.groupby(["artist", "title"]).agg({"value": "sum"}).reset_index()

# Drop the duplicate rows
deduped_df = df.drop_duplicates(subset=["artist", "title"])

# Merge the deduped dataframe with the grouped dataframe
merged_df = pd.merge(deduped_df, grouped_df, on=["artist", "title"])

# Drop the unnecessary columns
merged_df.drop(["value_x"], axis=1, inplace=True)
```

In diesem Code gruppieren wir den Datenrahmen zunächst nach den Spalten "Künstler" und "Titel" mit der groupby-Methode und verwenden dann die agg-Methode, um die Spalte "Werte" innerhalb jeder Gruppe zu summieren. Außerdem setzen wir den Index zurück, um den gruppierten Datenrahmen wieder in einen normalen Datenrahmen zu konvertieren.

Als Nächstes werden die doppelten Zeilen aus dem ursprünglichen Datenrahmen mit der drop_duplicates-Methode entfernt, wobei die Spalten "artist" und "title" als Teilmenge der zu berücksichtigenden Spalten ausgewählt werden.

Anschließend führen wir den entduplizierten Datenrahmen mit dem gruppierten Datenrahmen zusammen, indem wir die merge-Methode mit "artist" und "title" als Verknüpfungsschlüssel verwenden.

Nach dem Zusammenführen aktualisieren wir die Spalte "Werte", indem wir die summierten Werte aus dem gruppierten Datenrahmen zu den Werten im ursprünglichen Datenrahmen hinzufügen.

Schließlich löschen wir die überflüssige Spalte "values_x".
