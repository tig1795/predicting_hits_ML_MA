# CSV-Files in eine CSV-File zusammenf√ºhren

```python
#import modules
import os
import glob
#list all csv files only
csv_files = glob.glob('../../data/CSV/*.{}'.format('csv'))
sorted(csv_files)
```

```python
df_concat = pd.concat([pd.read_csv(f) for f in csv_files ], ignore_index=True)
df_concat
```

```python
df_concat.sort_values('year')
```

```python
df_concat.shape
```

```python
df_concat.to_csv('../../data/CSV/1950_2015.csv', encoding='utf-8', index=False)
```
