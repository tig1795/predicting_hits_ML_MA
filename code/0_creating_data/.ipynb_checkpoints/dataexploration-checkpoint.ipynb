# Bedeutung der einzelnen Spalten

- title: [string],              # Titel des Liedes.       
- artist: [string],             # Künstler des Liedes.
- year: [int],                  # Jahr der Veröffentlichung des Liedes.
- pos: [int],                   # Position der Billboard's Top 100 für das entsprechende Jahr.
- lyrics: [string],             # Text des Liedes.
- tags: [string array],         # Genre-Tags, die mit dem Künstler des Liedes verbunden sind.
- sentiment: {"neg": [float],# Negativität in Verbindung mit Texten(zwischen 0-1, 1 ist 100% negativ).
        "neu": [float],# Neutralität in Verbindung mit Texten (zwischen 0-1, 1 ist 100% neutral).
        "pos": [float],# Positivität in Verbindung mit dem Text (zwischen 0-1, 1 ist 100% positiv).
        "compound": [float]},
- f_k_grade: [float],           # Flesch-Kincaid-Niveau des Textes.
- flesch_index: [float],        # Flesch-Punktzahl für die Lesefähigkeit.
- fog_index: [float],           # Gunning-Fog-Lesbarkeitsindex.
- difficult_words: [int],       # Anzahl der Wörter, die nicht auf der Dale-Chall-Liste der "leichten" Wörter stehen.
- num_syllables: [int],         # Anzahl der Silben in Liedtexten.
- num_words: [int],             # Anzahl der Wörter im Liedtext.
- num_lines: [int],             # Anzahl der Zeilen im Liedtext.
- num_dupes: [int]              # Anzahl der doppelten (sich wiederholenden) Zeilen im Liedtext.
}

# Imports

```python
import pandas as pd
import numpy as np
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
songs = pd.read_csv("../../data/CSV/1950_2015.csv")
songs.head(5)
```

# Anzahl der Artists und Songs, dazugehörige Verteilungen

```python
for i,row in songs.iterrows():
    c = Counter(row.lyrics.split())
    songs.at[i, 'number_of_tokens'] = sum(c.values())
    songs.at[i, 'number_of_types'] = int(len(c))
```

```python
artists = np.unique(songs.artist)
print(len(np.unique(songs.artist)))
# np.unique(songs.decades)
```

```python
mean ={}
for a in artists:
    mean[a] = int(np.mean(songs.number_of_tokens[songs.artist==a]))
```

```python
ax = sns.distplot(list(mean.values()), rug=True, kde=False)
ax.set_xlabel("song length")
ax.set_ylabel("occurence")
ax.set_title("Distribution of Songlength")
plt.savefig("../../images/dataexploration/songlength")
plt.show()
```

```python
songs.sort_values('year')
songs.head(5)
```

## Hinzufügen der Dekaden

```python
songs['decades'] = '0'
songs.head(2)
```

```python
s = songs.copy()
for i, row in s.iterrows():
    if row.year >=1950 and row.year < 1960:
        s.at[i, 'decades'] = '1950s'
    if row.year >=1960 and row.year < 1970:
        s.at[i, 'decades'] = '1960s'
    if row.year >=1970 and row.year < 1980:
        s.at[i, 'decades'] = '1970s'
    if row.year >=1980 and row.year < 1990:
        s.at[i, 'decades'] = '1980s'
    if row.year >=1990 and row.year < 2000:
        s.at[i, 'decades'] = '1990s'
    if row.year >=2000 and row.year < 2010:
        s.at[i, 'decades'] = '2000s'
    if row.year >=2010 and row.year < 2020:
        s.at[i, 'decades'] = '2010s'
```

```python
s.head(5)
```

```python
s.to_csv("../../data/CSV/1950_2015_plus_dates.csv")
```
