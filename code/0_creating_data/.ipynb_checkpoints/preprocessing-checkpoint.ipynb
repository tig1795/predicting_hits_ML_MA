# Clean data

```python
import pandas as pd

songs = pd.read_csv('../../data/CSV/1950_2015_plus_dates.csv', index_col=0)
```

```python
songs.head(5)
```

```python
for i, row in songs.iterrows():
    songs.at[i, 'lyrics'] = row.lyrics.replace(',',' ').replace('(',' ').replace(')',' ').replace('\n', '').replace("[Chorus]", "").lower()
songs
```

```python
songs.to_csv("../../data/CSV/1950_2015_plus_dates.csv")
```

# Stemming the Lyrics

```python
songs = pd.read_csv("../../data/CSV/1950_2015_plus_dates.csv", index_col=0)

songs.head(5)
```

```python
import spacy
import en_core_web_sm
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

sentence = songs.lyrics[0]

ps = PorterStemmer()

for i, row in songs.iterrows():
    words = word_tokenize(row.lyrics)
    sent = ""
    for word in words:
        sent = sent + " " + ps.stem(word)
    songs.at[i,"stemmed_text"] = sent
songs
```

```python
songs.to_csv("../../data/CSV/1950_2015_plus_dates.csv")
```

```python
import pandas as pd
import numpy as np 

import matplotlib.pyplot as plt
import seaborn as sns

import regex as re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import nltk
nltk.download('punkt')

import spacy
from spacy import displacy

import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
songs = pd.read_csv("../../data/CSV/1950_2015_plus_dates.csv", index_col=0)
songs["POS_tags"] = "0"
songs['POS'] = '0'
songs.head(5)
```

## Exclusive use of nouns, verbs and adjectives

```python
nlp = en_core_web_sm.load()
tokens = []
for i, row in songs.iterrows():
    doc = nlp(row.lyrics)
    for token in doc:
        d = {"text":token.text, "lemma":token.lemma_, "pos":token.pos_, "tag":token.tag_, "dep":token.dep_,"shape":token.shape_, "alpha":token.is_alpha, "stop":token.is_stop}
        tokens.append(d)
    pos = pd.DataFrame(data=tokens)
    tokens = []  
    x = list(pos[pos.tag=="JJ"].text) + list(pos[pos.tag=="NN"].text) + list(pos[pos.tag=="VBP"].text)
    x = ' '.join(x)
    songs.at[i,'POS'] = x
#     songs.at[i, 'nouns'] = len(pos[pos.tag=="NN"].text)
#     songs.at[i, 'adjectives'] = len(pos[pos.tag=="JJ"].text)
#     songs.at[i, 'verb'] = len(pos[pos.tag=="VBP"].text)    

songs
```

# Convert CSV to XLSX file

```python
import os
import glob
import csv
from xlsxwriter.workbook import Workbook


for csvfile in glob.glob(os.path.join('.', '../../data/CSV/*.csv')):
    workbook = Workbook(csvfile[:-4] + '.xlsx')
    worksheet = workbook.add_worksheet()
    with open(csvfile, 'rt', encoding='utf8') as f:
        reader = csv.reader(f)
        for r, row in enumerate(reader):
            for c, col in enumerate(row):
                worksheet.write(r, c, col)
    workbook.close()
```
