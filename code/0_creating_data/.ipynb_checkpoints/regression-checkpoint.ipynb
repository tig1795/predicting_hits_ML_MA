```python
# Sortieren des DataFrame nach Songtitel und Jahr

data = data.sort_values(by=['title', 'year'])
```

```python
# Erstellen einer neuen Spalte, um den berechneten Wert zu speichern.

data['value'] = 0.0
```

```python
# Erstellen von Variablen, um die vorherige Position und die Jahre eines Songs zu verfolgen.

prev_position = None
prev_years = 0
prev_song = None
```

```python
# Iterieren über den DataFrame und berechnen des Wertes für jeden Eintrag.

for index, row in data.iterrows():
    song = row['title']
    pos = row['pos']
    year = row['year']
    if song != prev_song:
        # # Variablen für einen neuen Song zurücksetzen
        prev_position = None
        prev_years = 0
    if pos == prev_position:
        prev_years += 1
    else:
        prev_years = 1
    value = 1 / pos * prev_years
    data.at[index, 'value'] = value
    prev_position = pos
    prev_song = song
```

Der Code funktioniert folgendermaßen:

1. Der Code lädt die Daten in einen Pandas DataFrame und sortiert sie nach Songtitel und Jahr.

2. Dem DataFrame wird eine neue Spalte namens "value" hinzugefügt, um den berechneten Wert für jeden Eintrag zu speichern.

3. Der Code durchläuft jeden Eintrag im DataFrame und berechnet den Wert auf der Grundlage der Chartposition und der Anzahl der Jahre, in denen sich der Titel auf dieser Position befand.

4. Wenn der Song die gleiche Chartposition wie im Vorjahr hat, wird die Anzahl der Jahre an dieser Position erhöht.

5. Der berechnete Wert wird zu der Spalte "value" für jeden Eintrag im DataFrame hinzugefügt.

# Checken, wie viele Lieder zwei Jahre oder mehr aufeinander Folgenden die gleiche Chart-Position haben.

```python
# Sortieren des Dataframes nach Jahren
charts = charts.sort_values(by='year')

# Erstellen einer neuen Spalte, die die Differenz in Jahren zwischen jeder Zeile und der vorherigen Zeile berechnet
charts['year_diff'] = charts['year'].diff()

# Erstellen einer neuen Spalte, die angibt, ob die Position in den Charts in mindestens zwei aufeinanderfolgenden Jahren gleich geblieben ist.
charts['same_pos_for_2_years'] = (charts['year_diff'] == 1) & (charts['pos'] == charts['pos'].shift())

# Erstellen einer neuen Spalte, die die Spalten Titel und Künstler kombiniert.
charts['title_artist'] = charts['title'] + ' - ' + charts['artist']

# Erstellen einer neuen Spalte, die anzeigt, ob der Titel und der Interpret dieselben sind wie vor einem Jahr.
charts['same_title_artist_as_last_year'] = (charts['title_artist'] == charts['title_artist'].shift()) & (charts['year_diff'] == 1)

# Filtern des Dataframes so, dass er nur Lieder enthält, die seit mindestens zwei aufeinander folgenden Jahren dieselbe Chartposition innehaben und denselben Titel und Interpreten wie vor einem Jahr haben.
result = charts[(charts.groupby(['pos'])['same_pos_for_2_years'].transform('any')) & (charts.groupby(['title_artist'])['same_title_artist_as_last_year'].transform('any'))]

# Auswählen der Spalten "Titel", "Künstler", "Position" und "Jahr" aus und löschen aller Duplikate.
result = result[['title', 'artist', 'pos', 'year']].drop_duplicates()
```

Nach der Erstellung der Spalte "same_pos_for_2_years", die angibt, ob die Position in den Charts in mindestens zwei aufeinanderfolgenden Jahren gleich geblieben ist, wird eine neue Spalte "title_artist" erstellt, die die Spalten "title" und "artist" kombiniert. Anschließend wird eine neue Spalte mit der Bezeichnung "same_title_artist_as_last_year" erstellt, um anzugeben, ob die Spalte "title_artist" dieselbe ist wie vor einem Jahr (d. h. ob der Songname und der Interpret mit denen von vor einem Jahr identisch sind).

Anschließend wird das Dataframe so gefiltert, dass es nur Lieder enthält, die mindestens zwei Jahre in Folge dieselbe Chartposition innehatten und denselben Titel und Interpreten wie vor einem Jahr haben, basierend auf den Werten in den Spalten "same_pos_for_2_years" und "same_title_artist_as_last_year". Schließlich werden die Spalten "Titel", "Künstler", "Position" und "Jahr" ausgewählt und löschen alle Duplikate, um zu vermeiden, dass derselbe Titel zweimal angezeigt wird, wenn er in mehreren aufeinanderfolgenden Jahren dieselbe Position innehatte.

# Suche nach der Stelle, wo es sich anbieten würde, in hits & non hits einzuteilen

## Clustering

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Wertereihe
column = charts.value

# Konvertieren der Series in ein Numpy-Array und umformen
X = column.to_numpy().reshape(-1, 1)

# Verwenden des K-means-Clustering, um die Werte in drei Gruppen einzuteilen
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# Abrufen der Cluster-Beschriftungen und Zentren
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Erstellen eines Streudiagramms der Datenpunkte
plt.scatter(X[:, 0], np.zeros(X.shape[0]), c=labels, cmap='viridis')

# Hinzufügen der Clusterschwerpunkte zum Diagramm
plt.scatter(centroids[:, 0], np.zeros(centroids.shape[0]), c='red', s=100, marker='X')

# Finden der beiden Punkte, an denen sich die drei Cluster treffen
dividing_points = np.sort([(centroids[i] + centroids[j])/2 for i in range(2) for j in range(i+1, 3)])

print("The dividing point 1 is:", dividing_points[0])
print("The dividing point 2 is:", dividing_points[1])

# Vertikale Linien an den Trennungspunkten hinzufügen
plt.axvline(x=dividing_points[0], color='black', linestyle='--')
plt.axvline(x=dividing_points[1], color='black', linestyle='--')

plt.show()
```

Es wird das k-means-Clustering verwendet, um die Werte in drei Gruppen zu gruppieren, und um ein Streudiagramm der Datenpunkte zu erstellen, das auf der Grundlage ihrer Clusterbeschriftungen farblich gekennzeichnet ist. Es werden ebenfalls die Clusterschwerpunkte in das Diagramm hinzugefügt. Schließlich werden die beiden Punkte ermittelt, an denen sich die drei Cluster treffen, und an diesen Punkten werden vertikale Linien eingezeichnet. Das Ergebnis ist ein Diagramm, das das Clustering-Ergebnis und die beiden Trennungspunkte zeigt.

Es wird np.sort() verwendet, um die beiden Teilungspunkte in aufsteigender Reihenfolge zu sortieren, bevor die vertikalen Linien hinzugefügt werden. Es können mit verschiedenen Darstellungsparametern experimentiert werden, um die Visualisierung nach bestimmten Wünschen zu gestalten.

## Quantile-Funktion

```python
import pandas as pd
import matplotlib.pyplot as plt

# Angabe der Werte-Spalte im Dataframe
column = charts.value

# Ermitteln der Quantile, die die Daten in drei gleiche Gruppen unterteilen.
quantiles = column.quantile([1/3, 2/3])

# Extrahieren der Werte am 1/3- und 2/3-Quantil als Teilungspunkte.
dividing_point1 = quantiles[1/3]
dividing_point2 = quantiles[2/3]

print(f"The dividing point is {dividing_point1}")
print(f"The dividing point is {dividing_point2}")

# Erstellen eines Histogramms der Daten mit den markierten Teilungspunkten.
plt.hist(column, bins=10, color='blue')
plt.axvline(dividing_point1, color='red', linestyle='--', linewidth=2)
plt.axvline(dividing_point2, color='green', linestyle='--', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Column Values with Dividing Point')

plt.show()
```

##  Mittelwert und Standardabweichung

```python
import pandas as pd
import matplotlib.pyplot as plt

# Angabe der Werte-Spalte im Dataframe.
column = charts.value

# Berechnen des Mittelwerst und der Standardabweichung der Daten.
mean = column.mean()
std = column.std()

# Definieren der Teilungspunkte als eine Standardabweichung vom Mittelwert entfernt.
dividing_point1 = mean - std
dividing_point2 = mean + std

# Prüfen, ob die erste Gruppe leer ist, und gegebenenfalls anpassen der Trennpunkte.
if dividing_point1 <= column.min():
    dividing_point1 = column.quantile(0.25)
    dividing_point2 = column.quantile(0.75)

# Erstellen einer booleschen Maske, um die Daten in drei Gruppen aufzuteilen.
group1_mask = column < dividing_point1
group2_mask = (column >= dividing_point1) & (column <= dividing_point2)
group3_mask = column > dividing_point2

# Verwenden der booleschen Maske, um die Daten in drei Gruppen aufzuteilen.
group1 = column[group1_mask]
group2 = column[group2_mask]
group3 = column[group3_mask]

# Ausgabe der Teilungspunkte
print("Dividing points: {}, {}".format(dividing_point1, dividing_point2))

# Erstellung des Histogramms der Daten mit den Teilungspunkten.
plt.hist(column, bins=10, alpha=0.5)
plt.axvline(x=dividing_point1, color='red', linestyle='--')
plt.axvline(x=dividing_point2, color='red', linestyle='--')
plt.title("Data with dividing points")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()

# Ausgabe der Gruppen
#print("Group 1:", group1.values)
#print("Group 2:", group2.values)
#print("Group 3:", group3.values)
```

Dieser Code berechnet zunächst den Mittelwert und die Standardabweichung der Daten. Anschließend werden die Teilungspunkte als eine Standardabweichung vom Mittelwert entfernt definiert, wobei jedoch geprüft wird, ob die erste Gruppe leer ist, indem der erste Teilungspunkt mit dem Mindestwert in den Daten verglichen wird. Wenn der erste Teilungspunkt kleiner oder gleich dem Minimalwert in den Daten ist, werden die Teilungspunkte so angepasst, dass sie das 25. und 75. Perzentil der Daten sind, indem die quantile()-Methode von pandas series verwendet wird. Wenn die erste Gruppe leer ist, werden die Teilungspunkte auf diese Weise angepasst, um sicherzustellen, dass die erste Gruppe nicht leer ist.
