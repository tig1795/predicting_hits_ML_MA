```python
import pandas as pd

df = pd.read_csv("../../data/CSV/1950_2015_plus_dates_ohne_1950s_2.0.csv")
df.head(5)
```

```python
df.shape
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.head(5)
```

```python
from collections import Counter

# c = Counter(songs.artist)
c = Counter(df.decades)
c
```

```python
df = df.loc[:, ~df.columns.str.contains('^index')]
```

```python
df.head(5)
```

```python
df.shape
```

```python
df['Hit'] = '0'
df.head(2)
```

```python
s = df.copy()
for i, row in s.iterrows():
    if row.pos >=1 and row.pos <= 20:
        s.at[i, 'Hit'] = 'True'
    if row.pos >=80 and row.pos <= 100:
        s.at[i, 'Hit'] = 'False'
    if row.pos >=21 and row.pos <= 79:
        s.at[i, 'Hit'] = 'None'
```

```python
s.head(5)
```

```python
s.to_csv("../../data/CSV/dataset_lyrics_hits_non_hits_none.csv")
```

```python
s.shape
```

```python
data1 = s[s['Hit'] == 'True']
data2 = s[s['Hit'] == 'False']

sample = pd.concat([data1, data2], axis=0)
```
