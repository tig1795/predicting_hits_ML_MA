```python
pd.reset_option("max_rows")
```

```python
selected_columns = df_hits_ursprünglich[['artist', 'title']]
selected_columns
```

```python
selected_columns[selected_columns['title'] == "All That She Wants"]
```

```python
selected_columns.iloc[899:903]
```

```python
df_hits_spotify = pd.read_csv('../../../../Downloads/ma-hits.csv')
df_hits_spotify.head(5)
```

```python
selected_columns_sp = df_hits_spotify[['Artist Name(s)', 'Track Name']]
selected_columns_sp.head(5)
```

```python
# merge df2 into df1 based on their indexes
result_hits = selected_columns.merge(selected_columns_sp, left_index=True, right_index=True)
```

```python
result_hits.head(5)
```

```python
result_hits.shape
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# Create a copy of the DataFrame to avoid modifying the original data
df_copy = result_hits.copy()

# Define a function to apply styling to rows with artist name differences
def highlight_diff(row):
    # Compare values of 'artist' with 'Artist Name(s)'
    artist_diff = row['title'] != row['Track Name']
    
    # Apply styling to the row if artist names are different
    return ['background-color: yellow' if artist_diff else '' for _ in row]

# Apply the function to the DataFrame using the Styler class
styled_df = df_copy.style.apply(highlight_diff, axis=1)

# Display the styled DataFrame
styled_df
```
